```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

### Pie Chart : 
 A pie chart is a circular statistical graphic that is divided into slices to represent data. Each slice in a pie chart corresponds to a specific category or data point, and the size of each slice is proportional to the quantity it represents. Pie charts are commonly used to display the distribution or composition of a whole data set. Why we use pie chats beacaues :  
1. Pie charts allow for a quick visual comparison of the relative sizes of different categories.
2. They are intuitive and easy for most people to understand.
3. They work well for displaying simple data distributions.

```python
crypto_df = pd.DataFrame(data = {'allocation %' :[20,55,5,27,3]},index = ['BTC','ETH','LTC','XRP','ADA'])
print(crypto_df)
```

```python
crypto_df.plot.pie(y = 'allocation %',figsize = (6,6))
plt.title("Crypto portfolio pie chart")
```

```python
XRP = 60
BTC,LTC,ADA,ETH = (100-XRP)/4,(100-XRP)/4,(100-XRP)/4,(100-XRP)/4

rypto_df = pd.DataFrame(data={'allocation %':[XRP,BTC,LTC,ADA,ETH]},index = [XRP,BTC,LTC,ADA,ETH])
rypto_df.plot.pie(y='allocation %' , figsize =(6,6),explode = (0.1,0.05,0.05,0.05,0.05),labels=crypto_df.index,wedgeprops={'linewidth': 0.08, 'edgecolor': 'black'})
```

### Line plots : 

A line chart, also known as a line graph, is a type of data visualization that displays data points connected by straight lines. Line charts are commonly used to show trends over time or to visualize the relationship between two continuous variables. They are particularly effective for displaying data that changes continuously, such as stock prices, temperature variations, population growth, and more. And why we use line graphs because :
1. Line charts are excellent for showing trends and patterns in data over time or across a continuous range.
2. You can easily compare multiple data sets by plotting multiple lines on the same chart.
3. Line charts are simple to understand and provide a clear visual representation of data changes.

```python
Data1=pd.read_csv("crypto_daily_prices.csv")
Data1.head()
```

```python
Data1.plot(x='Date',y=['BTC','ETH','ADA'],figsize=(14,6),linewidth=2)
plt.ylabel("price $")
plt.title("Crypto price run")
#plt.legend("upper left")
plt.grid()
```

### Scatter plot :
A scatter plot is a type of data visualization that uses dots to represent the values obtained for two different variables. Each dot on the scatter plot represents a single data point, and the position of the dot on the plot corresponds to the values of the two variables. <b>Scatter plots are useful for identifying relationships or patterns between two continuous variables and for detecting the presence of outliers in data.</b> and scatter plots are usefull for :        
1. Scatter plots are effective for identifying relationships, correlations, or patterns between two continuous variables.
2. They can help in detecting outliers or unusual data points that don't follow the general trend.
3. Scatter plots provide insight into the distribution of data points.

```python
# if we want to get relation ship between two crypto currencies we can use sctter plot
Data1.plot.scatter(x='BTC',y='ETH',figsize=(8,8),color='r',alpha = 0.5)
```

### Histograme :

A histogram is a graphical representation that displays the distribution of a dataset. It groups data into intervals, or "bins," and shows the frequency or count of observations that fall into each bin. <b>Histograms are useful for understanding the underlying data distribution, identifying patterns, and visualizing the spread and shape of the data.</b> and why histograsms are usefull because:
1.  Histograms provide a clear visual representation of the distribution of a dataset, showing where the data is concentrated and how it spreads out.
2. Histograms can reveal patterns, gaps, clusters, and outliers in the data.
3. They provide a concise summary of the data's central tendency and variability.

```python
Data2 = pd.read_csv('crypto_daily_returns.csv')
Data2.head()
```

```python
plt.figure(figsize=(8,8))
Data2["BTC"].plot.hist(bins=25,color = 'blue',alpha= 0.7)
plt.ylabel("Probability")
plt.grid()
plt.title(f"AVerage : {Data2['BTC'].mean()} and Standard deviation : {Data2['BTC'].std()} ")
```

```python
plt.figure(figsize = (8,8))
Data2[['BTC',"ETH"]].plot.hist(bins=60, subplots = True , alpha = 0.7,grid=True,label=['BTC', 'ETH'])
# plt.grid()
#plt.ylabel("Probability Distribution")
```

```python
plt.figure(figsize = (10,8))
Data2[['BTC',"ETH"]].plot.hist(bins=60 , alpha = 0.7,grid=True,label=['BTC', 'ETH'])
plt.ylabel("Probability")
plt.title(f"BTC : AVerage : {Data2['BTC'].mean()} and Standard deviation : {Data2['BTC'].std()} \n ETH : AVerage : {Data2['ETH'].mean()} and Standard deviation : {Data2['ETH'].std()} ")
```

#### Displot :
Displot combines matplotlib histogram function with kdeplot(kernel density estimate). KDE plot is used to plot the probabilty density od a continuous variable.

```python
sb.distplot(x=Data3['mean radius'],bins=30,color = 'red')
```

```python
class_0 = Data3[(Data3["target"] == 0)]
class_1 = Data3[(Data3["target"] == 1)]
fig=plt.figure(figsize =(15,8))
plot= fig.add_subplot(121)
sb.distplot(class_0['mean radius'],bins=50)
plt.title("Class : target = 0")
plot= fig.add_subplot(122)
sb.distplot(class_1['mean radius'],bins=50)
plt.title("Class : target = 1")
```
