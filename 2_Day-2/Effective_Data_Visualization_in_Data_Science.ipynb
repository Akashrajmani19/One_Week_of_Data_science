```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

### Pie Chart : 
<pre> A pie chart is a circular statistical graphic that is divided into slices to represent data. Each slice in a pie chart corresponds to a specific category or data point, and the size of each slice is proportional to the quantity it represents. Pie charts are commonly used to display the distribution or composition of a whole data set. Why we use pie chats beacaues :  </pre>
1. Pie charts allow for a quick visual comparison of the relative sizes of different categories.
2. They are intuitive and easy for most people to understand.
3. They work well for displaying simple data distributions.

```python
crypto_df = pd.DataFrame(data = {'allocation %' :[20,55,5,27,3]},index = ['BTC','ETH','LTC','XRP','ADA'])
print(crypto_df)
```

```python
crypto_df.plot.pie(y = 'allocation %',figsize = (6,6))
plt.title("Crypto portfolio pie chart")
```

```python
XRP = 60
BTC,LTC,ADA,ETH = (100-XRP)/4,(100-XRP)/4,(100-XRP)/4,(100-XRP)/4

rypto_df = pd.DataFrame(data={'allocation %':[XRP,BTC,LTC,ADA,ETH]},index = [XRP,BTC,LTC,ADA,ETH])
rypto_df.plot.pie(y='allocation %' , figsize =(6,6),explode = (0.1,0.05,0,0,0))
```
