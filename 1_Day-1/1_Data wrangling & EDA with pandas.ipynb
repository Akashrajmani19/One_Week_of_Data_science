### Dealing with missing datain our table

```python
# Let's locate rows that have Null values
# Human_Res.isnull()  gives a table where the data is missing it will show false else it will show True
# To get sum of all missing values ccolumns wise we need to use sum()
Human_Res.isnull().sum()
```

```python
# Drop any row that contains a Null value
Human_Res.dropna(how='any' , inplace = False) # will drop every row in which any data is missing and "inplace = True"
                                             # will apply this chnages on original data
```

We can also put man,mod or median on the place of missing values and this is also a very effective way to deal with missing value:

```python
mean_value= Human_Res['MonthlyIncome'].mean()
Human_Res['MonthlyIncome'] = Human_Res['MonthlyIncome'].fillna(mean_value)
```

```python
Human_Res['MonthlyIncome'].isnull().sum()
```

Use the median value to fill out missing data in 'monthly rate':

```python
median_value=Human_Res['MonthlyRate'].median()
Human_Res['MonthlyRate'] = Human_Res['MonthlyRate'].fillna(median_value)
Human_Res['MonthlyIncome'].isnull().sum()
```
