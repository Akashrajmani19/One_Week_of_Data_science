### Dealing with missing data in our table

```python
# Let's locate rows that have Null values
# Human_Res.isnull()  gives a table where the data is missing it will show false else it will show True
# To get sum of all missing values ccolumns wise we need to use sum()
Human_Res.isnull().sum()
```

```python
# Drop any row that contains a Null value
Human_Res.dropna(how='any' , inplace = False) # will drop every row in which any data is missing and "inplace = True"
                                             # will apply this chnages on original data
```

We can also put man,mod or median on the place of missing values and this is also a very effective way to deal with missing value:

```python
mean_value= Human_Res['MonthlyIncome'].mean()
Human_Res['MonthlyIncome'] = Human_Res['MonthlyIncome'].fillna(mean_value)
```

```python
Human_Res['MonthlyIncome'].isnull().sum()
```

Use the median value to fill out missing data in 'monthly rate':

```python
median_value=Human_Res['MonthlyRate'].median()
Human_Res['MonthlyRate'] = Human_Res['MonthlyRate'].fillna(median_value)
Human_Res['MonthlyIncome'].isnull().sum()
```

### Feature Scaling : 
<pre>Feature Scaling is an important step to take prior to training of machine learning models to ensure that features are within the same scale. As eaxample intrest rate nd employment score are at diffrent scale. This will result in one eature dominating the other feature.</pre>            
### Normalization :
<pre>Normalization is conducted to make feature values renge from 0 to 1.</pre>
<pre>               x'=(x-min(x))/(max(x)-min(x))</pre>

### Standardization 
<pre>Standardization is conducted to tansform the data to have a "mean" of zero and "standard deviation" of 1. Standardization is also known as Z-score normalization in which properies will have the behaviour oa a standard normal distribution.</pre>       
<pre>             z = {x-avg(x)} / {standard deviation}</pre>


```python
Human_Res = pd.read_csv("Human_Resources.csv")
```

```python
Human_Res["Age"].values
```

```python
# Narmalization is conducted to make feature values range from 0 to 1.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Human_Res['Age'] = scaler.fit_transform(Human_Res['Age'].values.reshape(-1,1))
```

```python
Human_Res['Age'].describe()
```

```python
# Standardization is conducted to tansform the data to have a "mean" of zero and "standard deviation" of 1.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Human_Res["DailyRate"] =scaler.fit_transform(Human_Res["DailyRate"].values.reshape(-1,1))
```

### Pandas with functions 

```python
Data = pd.read_csv("Human_Resources.csv")
Data["DailyRate"].describe()
```

```python
# if daily rate has increased by 10% Then we can do this changes using function in very easy way
# Assuming "Data" is the DataFrame containing the balances and "increment" is 10%
def DailyRate(balance,increment):
    return balance*(1 + (increment/100))
Data["DailyRate"] = Data["DailyRate"].apply(lambda x: DailyRate(x, increment=10))
Data["DailyRate"].describe()
```

```python
# Let's define a function that doubles the DailyRate and add 100$
Data = pd.read_csv("Human_Resources.csv")
print(Data['DailyRate'].describe())
def modification(DailyRate):
    return DailyRate*2 + 100
Data['DailyRate'] = Data['DailyRate'].apply(modification)
print(Data['DailyRate'].describe())
```
